```python
import numpy as np
import pandas as pd
import pickle
```

```python
with open('country_to_int.pickle', 'rb') as handle:
    country_to_int = pickle.load(handle)
    
int_to_country = dict([[v,k] for k,v in country_to_int.items()])
```

```python
top_raw = pd.read_excel('Trade_gas.xlsx',sheet_name='Top')
top_raw.columns = top_raw.loc[0,:]+' '+top_raw.loc[1,:]
top_df = top_raw.drop([0,1], axis=0)
columns_of_interest = top_df.filter(regex='Country').columns.tolist()
for i in columns_of_interest:
    top_df[i] = top_df[i].map(int_to_country)
```

```python
top_df
```

```python
bottom_raw = pd.read_excel('Trade_gas.xlsx',sheet_name='Bottom')
bottom_raw.columns = bottom_raw.loc[0,:]+' '+bottom_raw.loc[1,:]
bottom_df = bottom_raw.drop([0,1], axis=0)
columns_of_interest = bottom_df.filter(regex='Country').columns.tolist()
for i in columns_of_interest:
    bottom_df[i] = bottom_df[i].map(int_to_country)
```

```python
bottom_df
```

```python
with pd.ExcelWriter('trade_edge_output.xlsx') as writer:  # doctest: +SKIP
    top_df.to_excel(writer, sheet_name='Top')
    bottom_df.to_excel(writer, sheet_name='Bottom')
```
