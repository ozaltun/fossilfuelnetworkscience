```python
USRankOilPR = list()
for i in range(2000,2016):   
    oilWeightPR, oilValuePR = pageRank(oilTrades_Total, i)
    country, rank =zip(*oilValuePR) 
    indx = country.index('USA')
    USRankOilPR.append((i,indx+1))

    
USRankGasPR = list()
for i in range(2000,2016):   
    gasWeightPR, gasValuePR = pageRank(gasTrades_Total, i)
    country, rank =zip(*gasValuePR) 
    indx = country.index('USA')
    USRankGasPR.append((i,indx+1))
    

USRankCoalPR = list()
for i in range(2000,2016):   
    coalWeightPR, coalValuePR = pageRank(coalTrades_Total, i)
    country, rank =zip(*coalValuePR) 
    indx = country.index('USA')
    USRankCoalPR.append((i,indx+1))
    

USRankTotalPR = list()
for i in range(2000,2016):   
    totWeightPR, totValuePR = pageRank(trades_Total, i)
    country, rank =zip(*totValuePR) 
    indx = country.index('USA')
    USRankTotalPR.append((i,indx+1))


year, rank0 = zip(*USRankOilPR)
plt.scatter(year, rank0, label='Oil')

year, rank0 = zip(*USRankGasPR)
plt.scatter(year, rank0, label='Gas')

year, rank0 = zip(*USRankCoalPR)
plt.scatter(year, rank0, label='Coal')

year, rank0 = zip(*USRankTotalPR)
plt.scatter(year, rank0, label='Total')

plt.legend()
plt.xlabel('Year')
plt.ylabel('Ranking of US')
plt.title('US PageRank Ranking')
plt.show()
plt.close()
```
